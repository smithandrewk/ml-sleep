```python
import torch
import os
from torch.utils.data import TensorDataset,ConcatDataset,DataLoader
import matplotlib.pyplot as plt
from torch import nn
from torch.nn.functional import relu
from tqdm import tqdm
from sklearn.preprocessing import RobustScaler
import seaborn as sns

device = 'cuda'
path_to_pt_ekyn = f'../pt_ekyn'
path_to_pt_snezana_mice = f'../pt_snezana_mice'

ekyn_ids = sorted(set([recording_filename.split('_')[0] for recording_filename in os.listdir(path_to_pt_ekyn)]))
snezana_mice_ids = sorted(set([recording_filename.split('.')[0] for recording_filename in os.listdir(path_to_pt_snezana_mice)]))
print(len(ekyn_ids),ekyn_ids)
print(len(snezana_mice_ids),snezana_mice_ids)

def load_ekyn(id,condition):
    X,y = torch.load(f'{path_to_pt_ekyn}/{id}_{condition}.pt',weights_only=False)
    return X,y
def load_snezana_mice(id):
    X,y = torch.load(f'{path_to_pt_snezana_mice}/{id}.pt',weights_only=False)
    return X,y
```

```python
id = ekyn_ids[0]

colors = [
    '#007AFF',  # Original blue
    '#FF9500',  # Original orange
    '#FF2D55',  # Bright red-pink
    '#5856D6',  # Deep purple
    '#00C7BE',  # Teal
    '#FFCC00',  # Golden yellow
    '#34C759',  # Fresh green
    '#AF52DE',  # Magenta
    '#FF3B30',  # Red
    '#5AC8FA',  # Light blue
    '#4CD964',  # Lime green
    '#FF9122',  # Deep orange
    '#1F2A44',  # Dark slate
    '#FFD60A',  # Bright yellow
    '#8E8E93',  # Gray
    '#32ADE6'   # Sky blue
]
linestyles = ['-','--']

plt.figure(figsize=(16,9))

for i,id in enumerate(ekyn_ids[:2]):
    for j,condition in enumerate(['PF','Vehicle']):
        X,y = load_ekyn(id,condition)
        sns.kdeplot(X.std(dim=1),cumulative=True,linewidth=2,color=colors[i],label=f'{id}_{condition}',linestyle=linestyles[j]);

designations = ['HET','KO','WK','WT']

for i,id in enumerate(snezana_mice_ids[:2]):
    for j,d in enumerate(designations):
        if d in id:
            break
    X,y = load_snezana_mice(id)
    sns.kdeplot(X.std(dim=1),cumulative=True,linewidth=2,label=f'{id}',color=colors[j]);

plt.xlim([0,.0002])
plt.legend()
```
